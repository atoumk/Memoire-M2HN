```python
# Date : float --> str
filtre_float = df_rmn['date'].apply(lambda x: isinstance(x, float) and pd.notnull(x))



filtre_int = df_rmn['date'].apply(lambda x: isinstance(x, int) and pd.notnull(x))



df_rmn.loc[filtre_float, 'date'] = df_rmn.loc[filtre_float, 'date'].astype(int).astype(str)
df_rmn.loc[filtre_int, 'date'] = df_rmn.loc[filtre_int, 'date'].astype(str)


df_rmn['date'] = df_rmn['date'].fillna('')
```

```python
# Données RMN : chaque lien -> une image et non un objet. Regrouper lignes pour un même objet.
df_rmn = df_rmn.groupby(['num_inventaire', 'localisation'], as_index=False).agg(lambda x: list(x))




for col in df_rmn.columns:
    if col not in ['num_inventaire', 'localisation']:
        df_rmn[col] = df_rmn[col].apply(lambda x: x[0] if len(set(x)) == 1 else x)
```

```python
# Localisation -> Vile & Institution
df_rmn['localisation'] = df_rmn['localisation'].str.replace('États-Unis,', '')
df_rmn['localisation'] = df_rmn['localisation'].str.replace('Royaume-Uni,', '') # Normaliser au format "ville, instution"
df_rmn['ville'] = df_rmn['localisation'].apply(lambda x: x.split(',')[0])
df_rmn['institution'] = df_rmn['localisation'].apply(lambda x: ','.join(x.split(',')[1:]).strip())
```

```python
def str_to_list(elem):
    if isinstance(elem, str) and elem.startswith('[') and elem.endswith(']'):
        try:
            return ast.literal_eval(elem)
        except (ValueError, SyntaxError):
            return elem
    return elem
```

```python
def create_id(df, mus_col_name, num_col_name, dico_mus_id, dico_chars):
    def clean_num_inv(text, dico):
        for key, value in dico.items():
            if key in text:
                text = text.replace(key, value)
        return text
    
    musee = df[mus_col_name].astype(str).str.strip()
    musee_id = musee.apply(lambda x: dico_mus_id.get(x, x))  # apply dictionary mapping
    num_inv = df[num_col_name].astype(str).str.strip()
    num_id = num_inv.apply(lambda x: clean_num_inv(x, dico_chars))  # apply cleaning
    
    # Concatenate to create object_id
    df['object_id'] = musee_id + '_' + num_id
```

# Enregistrer images

```python
def save_img (url_img, output_dir, object_id, timer=3) :
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/115.0.0.0 Safari/537.36"
        )
        }

    img_data = requests.get(url_img, headers=headers).content

    output_file = os.path.join(output_dir, object_id)    
    with open(output_file, 'wb') as f : 
        f.write(img_data)
    time.sleep(timer)
    return output_file
```
